LINEAR REGRESSION

```python
import numpy as np
import pandas as pd

X = np.array([[1, 2], [2, 4], [3, 5], [4, 4], [5, 5]])
y = np.array([3, 5, 7, 6, 9])

x_mean = np.mean(X, axis=0)
y_mean = np.mean(y)

numerator = np.sum((X[:, 0] - x_mean[0]) * (y - y_mean)) + np.sum((X[:, 1] - x_mean[1]) * (y - y_mean))
denominator = np.sum((X[:, 0] - x_mean[0])**2) + np.sum((X[:, 1] - x_mean[1])**2)
beta1 = numerator / denominator
beta0 = y_mean - beta1 * x_mean[0] - beta1 * x_mean[1]

results = pd.DataFrame({'Parameter': ['Intercept', 'Slope'], 'Value': [beta0, beta1]})

print(results)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

X = np.array([[1, 2], [2, 4], [3, 5], [4, 4], [5, 5]])
y = np.array([3, 5, 7, 6, 9])

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Target'] = y
df['Predicted'] = y_pred

sns.scatterplot(x='Feature1', y='Target', data=df)
plt.plot(df['Feature1'], df['Predicted'], color='red', label='Regression Line')
plt.title('Scatter Plot with Regression Line')
plt.legend()
plt.show()
```

GRADIENT DESCENT METHOD

```python
import numpy as np

X = np.array([1, 2, 3])
y = np.array([2, 4, 6])

m = 0
b = 0

learning_rate = 0.01
epochs = 10

n = len(X)

print("Epoch | Slope (m)    | Intercept (b)")
for epoch in range(epochs):
    y_pred = m * X + b

    dm = -(2/n) * np.sum(X * (y - y_pred))
    db = -(2/n) * np.sum(y - y_pred)

    m = m - learning_rate * dm
    b = b - learning_rate * db
    print(f"{epoch+1:>5} | {m:>12.4f} | {b:>14.4f}")
```
